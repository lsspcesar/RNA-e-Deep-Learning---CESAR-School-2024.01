# Regressão Linear com PyTorch

## Objetivos

Este exemplo utiliza regressão linear para estimar o comprimento das sépalas da íris a partir do seu comprimento das pétalas.
Utiliza-se:
- a função de perda MSE do PyTorch,
- treinamento dos parâmetros via gradiente descendente usando o otimizador.
- A rede é criada com uma camada nn.Linear()

## Importação dos pacotes

````python
%matplotlib inline
import torch
from torch import nn, optim
from torch.autograd import Variable
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

torch.manual_seed(1234)
````

## Leitura dos dados

````python
iris = load_iris()
data = iris.data[iris.target==1,::2]  # comprimento das sépalas e pétalas, indices 0 e 2

x_train = data[:,0:1].astype(np.float32)
y_train = data[:,1:2].astype(np.float32)

n_samples = x_train.shape[0]
print('x_train.shape:',x_train.shape, x_train.dtype)
print('y_train.shape:',y_train.shape, y_train.dtype)

print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
````

### Normalização dos dados

````python
x_train -= x_train.min()
x_train /= x_train.max()
y_train -= y_train.min()
y_train /= y_train.max()
````

````python
x_train_bias = np.hstack([np.ones(shape=(n_samples,1)), x_train])
````

````python
x_train_bias = torch.FloatTensor(x_train_bias)
y_train      = torch.FloatTensor(y_train)
````

## Criação do modelo da rede

````python
model = torch.nn.Linear(2, 1, bias=False)
````

### Verificando a inicialização dos parâmetros

````python
model.weight.data = torch.zeros(1,2)
torch.nn.init.uniform(model.weight.data, -0.1, 0.1)
model.weight.data
````

### Testando o predict da rede

````python
model(Variable(torch.ones((5,2))))
````

## Treinamento

### Definindo função de perda e otimizador

````python
criterion = nn.MSELoss() #
optimizer = optim.SGD(model.parameters(), lr=0.3) # Usando aqui o SGd. O Pytorch tem vários outros. lr - learning reating - atualiza os pesos automaticamente
````

### Laço de treinamento

````python
num_epochs = 120
w0_list = []
w1_list = []
for epoch in range(num_epochs):
    inputs = Variable(x_train_bias)
    target = Variable(y_train)

    # forward - predict
    out = model(inputs)

    w0_list.append(model.weight.data[0][0].item())
    w1_list.append(model.weight.data[0][1].item())

    #loss cálculo da função de perda
    loss = criterion(out, target) # usando os parâmetros

    # backward e otimizador (vai substituir o cálculo manual do gradiente)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # verbose
    if (epoch+1) % 20 == 0:
        print('Epoch[{}/{}], loss: {:.6f}'
              .format(epoch+1, num_epochs, loss.data))
````

````python
best_epoch_weight = None
best_epoch_number = 0
best_epoch_loss = float('inf')  # Inicializa com infinito para garantir que qualquer perda será menor
num_epochs = 120

# Lista para armazenar os pesos de cada época
w0_list = []
w1_list = []

for epoch in range(num_epochs):
    inputs = Variable(x_train_bias)  # Dados de entrada
    target = Variable(y_train)  # Rótulos de saída

    # Forward - previsão
    out = model(inputs)

    # Armazenando os pesos para visualização
    w0_list.append(model.weight.data[0][0].item())
    w1_list.append(model.weight.data[0][1].item())

    # Cálculo da função de perda
    loss = criterion(out, target)  # Calcula a perda usando a saída e o target

    # Salvar o melhor modelo (menor perda)
    if loss.item() < best_epoch_loss:  # Compare com a perda mínima
        best_epoch_loss = loss.item()  # Atualiza a melhor perda
        best_epoch_weight = model.state_dict()  # Salva os pesos do melhor modelo
        best_epoch_number = epoch + 1  # Armazena a época com a melhor perda

    # Backward - cálculo do gradiente e atualização dos parâmetros
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Imprime a perda a cada 20 épocas
    if (epoch + 1) % 20 == 0:
        print(f'Epoch[{epoch+1}/{num_epochs}], Loss: {loss.item():.6f}')

# Após o treinamento, podemos acessar o melhor modelo:
print(f"Melhor modelo encontrado na época {best_epoch_number} com perda {best_epoch_loss:.6f}")

# Opcional: Visualizar pesos em relação às épocas
plt.figure(figsize=(10, 5))
plt.plot(w0_list, label="Peso w0 (primeiro peso)")
plt.plot(w1_list, label="Peso w1 (segundo peso)")
plt.legend()
plt.xlabel("Épocas")
plt.ylabel("Valor do peso")
plt.title("Evolução dos Pesos durante o Treinamento")
plt.show()
````

````python
print(best_epoch_number)
print(best_epoch_loss)
print(best_epoch_weight)
````

## Avaliação

````python
y_pred = model(Variable(x_train_bias))
plt.plot(x_train, y_train.numpy(), 'ro', label='Original data')
plt.plot(x_train, y_pred.data.numpy(), 'kx-', label='Fitting Line')
plt.show()
````
